#  Исследование результатов А/В-теста и поиск инсайтов

SollmaFin — международное мобильное приложение для онлайн-торговли финансовыми активами. Пользователи могут инвестировать деньги в акции, валюту, криптовалюту, биржевые фонды (ETF) и другие активы. Целевая аудитория приложения — начинающие инвесторы. Получить доступ к финансовым рынкам можно со смартфона либо через веб-версию. 

Приложение ориентировано на рынок Латинской Америки с акцентом на четыре страны: Мексику, Бразилию, Колумбию и Аргентину — там быстро растёт интерес к финтеху и инвестициям, а экономики очень разные. 

Появилась гипотеза о недостаточной финансовой грамотности пользователей: они не понимают, как работать с активами разной степени риска. Клиенты покупают активы с высоким риском и теряют деньги, что снижает вовлечённость.
Команда продукта решила обновить онбординг пользователей и добавить в него детальную информацию о различиях финансовых активов и связанных с ними рисках. Однако появилось опасение, что углублённый онбординг отпугнёт клиентов от пополнения депозита и более рискованных вложений. 

Чтобы разобраться, команда решила провести A/B-эксперимент. 

### А/В-эксперимент

В эксперименте участвовали новые пользователи, которые зарегистрировались в приложении со 2 по 15 июня 2025 года. 
Пользователей случайным образом разделили на две равные группы: 
- Контрольная группа проходила стандартный онбординг без обязательного обучения.
- Тестовая группа проходила обновлённый онбординг с подробной информацией об активах и связанных с ними рисках. 

После разделения активность пользователей анализировали в течение недели.

Цель эксперимента — оценить, как обновлённый онбординг влияет на поведение пользователей и их дальнейшую инвестиционную активность.

При внедрении новой фичи команда продукта выдвинула такие гипотезы:
- Гипотеза роста: обучающий онбординг помогает пользователям лучше понимать принципы инвестирования, поэтому они будут чаще открывать второй депозит.
- Гипотеза риска: информация о возможных потерях и высоких рисках отпугнёт некоторых новичков, особенно самых осторожных, что снизит конверсию в первый депозит.
- Дополнительная гипотеза: после нового онбординга пользователи, которые выбрали высокорискованные активы, будут чаще, чем раньше, возвращаться и открывать второй депозит. При старом онбординге пользователи часто покупали активы с высоким риском без понимания последствий. Это приводило к потерям и оттоку после первого депозита. 

Чтобы увидеть полную картину, недостаточно следить за метриками конверсии. Команда выбрала для эксперимента такой набор метрик: 
- Ключевая метрика — средняя сумма всех депозитов на одного пользователя (включая тех, кто установил приложение или открыл веб-версию).
- Барьерная метрика — конверсия из регистрации в первый депозит.
- Вспомогательная метрика 1 — конверсия из первого депозита во второй.
- Вспомогательная метрика 2 — средняя сумма всех депозитов на пользователя, который открыл хотя бы один депозит.

Ожидалось, что ключевая и барьерная метрики не упадут, а вспомогательные покажут значительный рост.

**Задача 2.1.** Анализ новых пользователей.
- Изучите динамику привлечения новых пользователей в приложение.

````python
df_hist['event_dt'] = pd.to_datetime(df_hist['event_ts'].dt.date)
# Фильтруем по названию события 'registration'
registrations = df_hist.copy()
#убираем дубликаты 
registrations = registrations.drop_duplicates(subset = 'user_id')

new_users_dynamic = registrations.groupby('first_dt')['user_id'].nunique().reset_index().rename(columns = {'user_id':'users_count'})
new_users_dynamic = new_users_dynamic.sort_values('first_dt')
new_users_dynamic['cumulative_users'] = new_users_dynamic['users_count'].cumsum()


#Строим график для новых пользователей по дням
plt.figure(figsize = (16,8))
plt.plot(new_users_dynamic['first_dt'], new_users_dynamic['users_count'], color = 'mediumslateblue', marker = 'o')
plt.title('Динамика новых пользователей по дням')
plt.xlabel('Дата')
plt.ylabel('Новые пользователи')
plt.axhline(y = new_users_dynamic['users_count'].mean(), color = 'black', linestyle='--',
            label=f'Среднее: {new_users_dynamic["users_count"].mean():.1f}')
plt.grid()
plt.legend()
plt.show()

#Кумулятивный график
plt.figure(figsize = (16,8))
plt.plot(new_users_dynamic['first_dt'], new_users_dynamic['cumulative_users'], color = 'mediumslateblue', marker = 'o')
plt.title('Динамика новых пользователей кумулятивно по дням')
plt.xlabel('Дата')
plt.ylabel('Общее количество новых пользователей')
plt.grid()
plt.show()
````

````python
# Создаем колонки для группировки
registrations['time_period'] = registrations['event_ts'].apply(
    lambda x: '00:00-12:00' if x.hour < 12 else '12:00-00:00'
)

registrations = registrations.drop_duplicates(subset=['user_id', 'first_dt', 'time_period'])

registrations_period = registrations.groupby(['first_dt', 'time_period'])['user_id'].nunique().unstack()
registrations_period = registrations_period.fillna(0)

plt.figure(figsize=(16, 8))

# График для 00:00-12:00
plt.plot(registrations_period.index, 
        registrations_period['00:00-12:00'], 
         color='mediumslateblue', 
         marker='o',
         label='00:00-12:00')

# График для 12:00-00:00
plt.plot(registrations_period.index, 
         registrations_period['12:00-00:00'], 
         color='navy', 
         marker='o',
         label='12:00-00:00')

plt.title('Динамика новых пользователей по полуденным периодам')
plt.xlabel('Дата')
plt.ylabel('Количество регистраций')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()

# Сортируем по дате для корректного расчета кумулятивной суммы
registrations_period = registrations_period.sort_index()

# Рассчитываем кумулятивные суммы
registrations_period['00:00-12:00_cumsum'] = registrations_period['00:00-12:00'].cumsum()
registrations_period['12:00-00:00_cumsum'] = registrations_period['12:00-00:00'].cumsum()

plt.figure(figsize = (16,8))
plt.plot(registrations_period.index, registrations_period['00:00-12:00_cumsum'], 
        color='mediumslateblue', label='00:00-12:00 (кумулятивно)')
plt.plot(registrations_period.index, registrations_period['12:00-00:00_cumsum'], 
        color='navy', label='12:00-00:00 (кумулятивно)')
plt.title('Кумулятивная динамика новых пользователей по периодам')
plt.xlabel('Дата')
plt.ylabel('Общее количество регистраций')
plt.grid(True, alpha=0.3)
plt.legend()

plt.tight_layout()
plt.show()
````

**Задача 2.2.** Анализ воронок событий.
- Постройте классическую и пошаговую воронки;
- Охарактеризуйте пользовательский путь в приложении и определите узкие места воронки. 


````python
#Рассчитываем данные воронок  
funnel_steps = ['install / open_web', 'introduction', 'registration', 'main_page', 
                'onboarding_complete', 'first_deposit', 'asset_purchase', 'second_deposit']

funnel_counts = df_hist.groupby('event_name')['user_id'].nunique().reindex(funnel_steps)

funnel_df = pd.DataFrame({
    'step':funnel_steps,
    'users':funnel_counts.values
})

funnel_df['conversion_from_first_%'] = round(funnel_df['users'] * 100 / funnel_df['users'].iloc[0], 2)
funnel_df['conversion_from_previous_%'] = round(funnel_df['users'] * 100 / funnel_df['users'].shift(1), 2)

display(funnel_df)
````

````python
# Визуализируем данные классической воронки
plt.figure(figsize = (12,6))

bars = plt.bar(funnel_df['step'],
       funnel_df['conversion_from_first_%'], 
       color = 'mediumslateblue', 
       edgecolor = 'navy')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.1f}%',
             ha='center', va='bottom',
             fontsize=10)
plt.title('Классическая воронка')
plt.ylabel('Конверсия от первого шага (%)')
plt.xlabel('Этапы воронки')
plt.xticks(rotation = 45)
plt.show()
````

````python
# Визуализируем данные воронки step-by-step
plt.figure(figsize = (12,6))

bars = plt.bar(funnel_df['step'][1:],
       funnel_df['conversion_from_previous_%'][1:], 
       color = 'mediumslateblue', 
       edgecolor = 'navy')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.1f}%',
             ha='center', va='bottom',
             fontsize=10)
plt.title('Классическая воронка')
plt.ylabel('Конверсия от предыдущего шага (%)')
plt.xlabel('Этапы воронки')
plt.xticks(rotation = 45)
plt.show()
````

**Задача 3.2.** Сравнение воронок событий.
- Сравните пользовательский путь новых пользователей в тестовой и контрольной группах.
- Постройте классическую и последовательную воронки, выполните визуализацию и сделайте промежуточный вывод. 

````python
#Рассчитываем данные воронок  
funnel_steps = ['install / open_web', 'introduction', 'registration', 'main_page', 
                'onboarding_complete', 'first_deposit', 'asset_purchase', 'second_deposit']
funnel_df = []
for group in df_abt['group'].unique():
    funnel_counts = df_abt[df_abt['group'] == group].groupby('event_name')['user_id'].nunique().reindex(funnel_steps)
    funnel = pd.DataFrame({
    'group':group,
    'step':funnel_steps,
    'users':funnel_counts.values
    })
    funnel['conversion_from_first_%'] = round(funnel['users'] * 100 / funnel['users'].iloc[0], 2)
    funnel['conversion_from_previous_%'] = round(funnel['users'] * 100 / funnel['users'].shift(1), 2)
    funnel_df.append(funnel)
    
funnel_df = pd.concat(funnel_df)
funnel_df
````

````python
#Строим классические воронки
for group in funnel_df['group'].unique():
    df_plot = funnel_df[funnel_df['group'] == group]
    plt.figure(figsize = (12,6))
    bars = plt.bar(df_plot['step'],
       df_plot['conversion_from_first_%'], 
       color = 'mediumslateblue', 
       edgecolor = 'navy')

    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height,
                 f'{height:.1f}%',
                 ha='center', va='bottom',
                 fontsize=10)
    plt.title(f'Классическая воронка {group} группы')
    plt.ylabel('Конверсия от первого шага (%)')
    plt.xlabel('Этапы воронки')
    plt.xticks(rotation = 45)
    plt.show()
````

Анализ воронки показывает стабильные результаты на первых четырех этапах в обеих группах с сопоставимыми показателями конверсии. Однако на этапе онбординга наблюдается заметное снижение эффективности в тестовой группе — разница составляет 8.2 процентных пункта. Анализ воронки показывает, что процент пользователей, совершивших первый депозит, практически одинаков в обеих группах (66.6% в контрольной против 65.4% в тестовой), что свидетельствует о равных стартовых условиях. При этом метрика повторного депозита демонстрирует значительный рост в тестовой группе: 21.3% против 13.5% в контрольной, что указывает на успешное воздействие экспериментальных изменений пользователей, несмотря на ухудшение показателей онбординга.

````python
#Строим воронки step-by-step
for group in funnel_df['group'].unique():
    df_plot = funnel_df[funnel_df['group'] == group]
    plt.figure(figsize = (12,6))
    bars = plt.bar(df_plot['step'][1:],
       df_plot['conversion_from_previous_%'][1:], 
       color = 'mediumslateblue', 
       edgecolor = 'navy')

    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height,
                 f'{height:.1f}%',
                 ha='center', va='bottom',
                 fontsize=10)
    plt.title(f'Step-by-step воронка {group} группы')
    plt.ylabel('Конверсия от предыдущего шага (%)')
    plt.xlabel('Этапы воронки')
    plt.xticks(rotation = 45)
    plt.show()
````

Step-by-step воронка контрольной группы демонстрирует несколько более высокие результаты на начальных этапах, особенно заметна разница в показателях онбординга — 97.6% против 88.3% в тестовой группе. Такое расхождение может сигнализировать о потенциальных сложностях с вовлечением пользователей в процесс адаптации. Однако на последующих этапах — первом депозите, покупке актива и открытии второго депозита — тестовая группа показывает лучшую динамику, компенсируя первоначальное отставание. Финальный этап воронки завершается с разницей в 7% между группами.

````python
#Собираем данные для построения графика
cr_df = []
for group in df_abt['group'].unique():
    first_deposit = df_abt[(df_abt['event_name'] == 'first_deposit') & (df_abt['group'] == group)]
    second_deposit = df_abt[(df_abt['event_name'] == 'second_deposit') & (df_abt['group'] == group)]

    first_risks = df_abt[df_abt['event_name'] == 'asset_purchase']
    first_deposits_risks = pd.merge(first_deposit['user_id'], first_risks[['user_id', 'risk_level']], how = 'inner', on = 'user_id')
    first_deposits_count = first_deposits_risks.groupby('risk_level')['user_id'].nunique().reset_index()

    second_deposits_risks = pd.merge(second_deposit['user_id'], first_risks[['user_id', 'risk_level']], how = 'left', on = 'user_id')
    second_deposits_count = second_deposits_risks.groupby('risk_level')['user_id'].nunique().reset_index()

    cr_deposits = pd.merge(first_deposits_count, second_deposits_count, how = 'inner', on = 'risk_level')
    cr_deposits['conversion_to_second'] = round(cr_deposits['user_id_y'] / cr_deposits['user_id_x'] * 100, 2)
    cr_deposits['group'] = group
    cr_deposits = cr_deposits.sort_values(by='conversion_to_second')
    cr_df.append(cr_deposits)
    
cr_df = pd.concat(cr_df)

cr_df = cr_df.rename(columns = {'user_id_x': 'first_deposit_users', 'user_id_y': 'second_deposit_users' })
cr_df = cr_df[['group', 'risk_level', 'first_deposit_users', 'second_deposit_users', 'conversion_to_second']]

cr_df
````

Это видно в обеих группах. Однако тестовая демонстрирует значительно более хорошие показатели - почти половина (46.0%) пользователей с высокорискованным активом открывают второй депозит, в то время как в контрольной группе этот показатель равен 27.7%. Также большое превосходство показывает конверсия тестовой группы, где актив был низкорисковым - 57.5% против 49.8%. Перейдем к рассчету статистической значимости изменений, чтобы принять или опровергнуть гипотезы.

***H0:*** Конверсия во второй депозит у пользователей с низким уровнем риска первого актива одинакова и в контрольной, и в тестовой группе.

***H1:*** Конверсии во второй депозит у пользователей с низким уровнем риска первого актива различаются в контрольной и тестовой группе.



***H0:*** Конверсия во второй депозит у пользователей со средним уровнем риска первого актива одинакова и в контрольной, и в тестовой группе.

***H1:*** Конверсии во второй депозит у пользователей со средним уровнем риска первого актива различаются в контрольной и тестовой группе.



***H0:*** Конверсия во второй депозит у пользователей с высоким уровнем риска первого актива одинакова и в контрольной, и в тестовой группе.

***H1:*** Конверсии во второй депозит у пользователей с высоким уровнем риска первого актива различаются в контрольной и тестовой группе.


````python
# Переходим к расчету статистической значимости изменений: используем z-test, так как сравниваем конверсии
results = []

for risk in cr_df['risk_level'].unique():
    risk_data = cr_df[cr_df['risk_level'] == risk]
    
    control = risk_data[risk_data['group'] == 'control']
    test = risk_data[risk_data['group'] == 'test']
    
    control_first_deposits = control['first_deposit_users'].values[0]
    control_second_deposits = control['second_deposit_users'].values[0]
    test_first_deposits = test['first_deposit_users'].values[0]
    test_second_deposits = test['second_deposit_users'].values[0]
    
    successes = [test_second_deposits, control_second_deposits]
    total = [test_first_deposits, control_first_deposits]
    
    z_stat, p_value = proportions_ztest(successes, total, alternative = 'two-sided')
    results.append({
            'risk_level': risk,
            'control_conversion': control['conversion_to_second'].values[0],
            'test_conversion': test['conversion_to_second'].values[0],
            'z_statistic': z_stat,
            'p_value': p_value,
            'significant': p_value < 0.05 
        })

results_df = pd.DataFrame(results)
pd.set_option('display.float_format', lambda x: '%.6f' % x)
results_df
````

Рассчет p-value показал, что есть статистически значимый эффект на пользователей, чей первый актив был высокорисковым и низкорисковым. Это значит, что две из трех альтернативных гипотез подтвердились и в этих группах изменения в онбординге действительно приводят к росту открытия второго депозита.

- Барьерная метрика — конверсия из регистрации в первый депозит.

````python
#Пишем функцию, которая рассчитает все метрики и соберет их в единую таблицу
def calculate_metric(df):
    results = []
    # Ключевая метрика: средняя сумма депозитов на всех пользователей
    total_deposits = df.groupby('group')['amount'].sum()
    total_users = df.groupby('group')['user_id'].nunique()
    avg_deposits = total_deposits / total_users
    absolute_lift = avg_deposits['test'] - avg_deposits['control']
    relative_lift = (avg_deposits['test'] - avg_deposits['control']) / avg_deposits['control'] * 100
    results.append({
        'Метрика': 'Средняя сумма депозитов на всех пользователей',
        'Контрольная группа': round(avg_deposits['control'], 2),
        'Тестовая группа': round(avg_deposits['test'], 2),
        'Абсолютная разница': round(absolute_lift, 2),
        'Относительная разница (%)': round(relative_lift, 2)
    })
    #Барьерная метрика — конверсия из регистрации в первый депозит
    registrated = df[df['event_name'] == 'registration'].groupby('group')['user_id'].nunique()
    first_deposit = df[df['event_name'] == 'first_deposit'].groupby('group')['user_id'].nunique()
    cr_to_first = first_deposit / registrated * 100
    absolute_lift = cr_to_first['test'] - cr_to_first['control']
    relative_lift = (cr_to_first['test'] - cr_to_first['control']) / cr_to_first['control'] * 100
    results.append({
        'Метрика': 'Конверсия из регистрации в первый депозит',
        'Контрольная группа': round(cr_to_first['control'], 2),
        'Тестовая группа': round(cr_to_first['test'], 2),
        'Абсолютная разница': round(absolute_lift, 2),
        'Относительная разница (%)': round(relative_lift, 2)
    })
    #Вспомогательная метрика 1 — конверсия из первого депозита во второй
    first_deposit = df[df['event_name'] == 'first_deposit'].groupby('group')['user_id'].nunique()
    second_deposit = df[df['event_name'] == 'second_deposit'].groupby('group')['user_id'].nunique()
    cr_to_second = second_deposit / first_deposit * 100
    absolute_lift = cr_to_second['test'] - cr_to_second['control']
    relative_lift = (cr_to_second['test'] - cr_to_second['control']) / cr_to_second['control'] * 100
    results.append({
        'Метрика': 'Конверсия из первого депозита во второй',
        'Контрольная группа': round(cr_to_second['control'], 2),
        'Тестовая группа': round(cr_to_second['test'], 2),
        'Абсолютная разница': round(absolute_lift, 2),
        'Относительная разница (%)': round(relative_lift, 2)
    })
    #Вспомогательная метрика 2 — средняя сумма всех депозитов на пользователя, который открыл хотя бы один депозит
    users_with_first = df[df['event_name'] == 'first_deposit']['user_id'].unique()
    total_deposits = df.groupby('group')['amount'].sum()
    total_users = df[df['user_id'].isin(users_with_first)].groupby('group')['user_id'].nunique()
    avg_deposits = total_deposits / total_users
    absolute_lift = avg_deposits['test'] - avg_deposits['control']
    relative_lift = (avg_deposits['test'] - avg_deposits['control']) / avg_deposits['control'] * 100
    results.append({
        'Метрика': 'Средняя сумма всех депозитов на пользователя, который открыл хотя бы один депозит',
        'Контрольная группа': round(avg_deposits['control'], 2),
        'Тестовая группа': round(avg_deposits['test'], 2),
        'Абсолютная разница': round(absolute_lift, 2),
        'Относительная разница (%)': round(relative_lift, 2)
    })
    return pd.DataFrame(results)

display(calculate_metric(df_abt))
````

 Средняя сумма депозитов на всех пользователей в тестовой группе чуть выше 92.46 против 90.24 в контрольной. Такой же незначительный рост показывает и средняя сумма депозитов на пользователя, который открыл хотя бы один депозит. Так же значительный прирост в 60% можно увидеть в тестовой группе по метрике конверсия из первого депозита во второй. Это важная динамика, так как именно эта конверсия беспокоила компанию. Однако в тестовой группе происходит небольшое снижение барьерной метрики -  конверсии из регистрации в первый депозит. Узнаем, значимы ли эти изменения статистически. 
 
Перейдем к ключевой метрике и составим две гипотезы. 

***H0: Средняя сумма всех депозитов на одного пользователя (включая тех, кто установил приложение или открыл веб-версию) одинакова и в контрольной, и в тестовой группе.***

***H1: Средняя сумма всех депозитов на одного пользователя (включая тех, кто установил приложение или открыл веб-версию) тестовой группе отлична от контрольной.***

````python
# Оцениваем статистическую значимость различий в ключевой метрике групп
#Подготовим данные для анализа
df_abt['event_dt'] = pd.to_datetime(df_abt['event_ts'].dt.date)

dates = sorted(df_abt['event_dt'].unique())
results = []

for date in dates:
    df_date = df_abt[df_abt['event_dt'] <= date]
    users_deposits = df_date.groupby(['user_id', 'group'])['amount'].sum().reset_index().fillna(0)
    arpu_control = users_deposits[users_deposits['group'] == 'control']['amount'].mean()
    arpu_test = users_deposits[users_deposits['group'] == 'test']['amount'].mean()
    control = users_deposits[users_deposits['group'] == 'control']['amount']
    test = users_deposits[users_deposits['group'] == 'test']['amount']
    _, p_value = ttest_ind(test, control, equal_var = False)
    
    results.append({
        'date':date,
        'arpu_control':arpu_control,
        'arpu_test':arpu_test,
        'p_value':p_value
    })
    
df_results = pd.DataFrame(results)
df_results
````

Гипотезы изменения барьерной метрики:

***H0: Конверсия из регистрации в первый депозит одинакова и в контрольной, и в тестовой группе.***

***H1: Конверсия из регистрации в первый депозит в тестовой группе отлична от конверсии в контрольной.***

````python
# Перейдем к оценке статистической значимости различий между группами, начнем с барьерной метрики
# Используем z-test
registrated = df_abt[df_abt['event_name'] == 'registration'].groupby('group')['user_id'].nunique().reset_index()
first_deposit = df_abt[df_abt['event_name'] == 'first_deposit'].groupby('group')['user_id'].nunique().reset_index()
reg_to_first = pd.merge(registrated, first_deposit, how = 'left', on = 'group')

control_registration = reg_to_first[reg_to_first['group'] == 'control']['user_id_x'].values[0]
test_registration = reg_to_first[reg_to_first['group'] == 'test']['user_id_x'].values[0]

control_first_deposit = reg_to_first[reg_to_first['group'] == 'control']['user_id_y'].values[0]
test_first_deposit = reg_to_first[reg_to_first['group'] == 'test']['user_id_y'].values[0]

successes = [test_first_deposit, control_first_deposit]
total = [test_registration, control_registration]

z_stat, p_value = proportions_ztest(successes, total, alternative = 'two-sided')
if p_value < 0.05:
    display(f'Значение p-value = {round(p_value, 4)}. Подтверждается альтернативная гипотеза. Изменения конверсии из регистрации в первый депозит статистически значимы')
else:
    display(f'Значение p-value = {round(p_value, 4)}. Подтверждается нулевая гипотеза. Изменения конверсии из регистрации в первый депозит статистически незначимы')
````

Гипотезы изменения вспомогательной метрики 1 - конверсия из первого депозита во второй.

***H0: Конверсия из первого депозита во второй одинакова и в контрольной, и в тестовой группе.***

***H1: Конверсия из первого депозита во второй в тестовой группе отлична от конверсии в контрольной.***

````python
#Проверим статистическую значимость конверсии из первого депозита во второй 
first_deposit = df_abt[df_abt['event_name'] == 'first_deposit'].groupby('group')['user_id'].nunique().reset_index()
second_deposit = df_abt[df_abt['event_name'] == 'second_deposit'].groupby('group')['user_id'].nunique().reset_index()
cr_to_second = pd.merge(first_deposit, second_deposit, how = 'left', on = 'group')

control_first = cr_to_second[cr_to_second['group'] == 'control']['user_id_x'].values[0]
test_first = cr_to_second[cr_to_second['group'] == 'test']['user_id_x'].values[0]

control_second = cr_to_second[cr_to_second['group'] == 'control']['user_id_y'].values[0]
test_second = cr_to_second[cr_to_second['group'] == 'test']['user_id_y'].values[0]

successes = [test_second, control_second]
total = [test_first, control_first]

z_stat, p_value = proportions_ztest(successes, total, alternative = 'two-sided')
if p_value < 0.05:
    display(f'Значение p-value = {round(p_value, 4)}. Подтверждается альтернативная гипотеза. Изменения конверсии из первого депозита во второй статистически значимы.')
else:
    display(f'Значение p-value = {round(p_value, 4)}. Подтверждается нулевая гипотеза. Изменения конверсии из первого депозита во второй статистически незначимы.')
````

Гипотезы изменения вспомогательной метрики 2 - средняя сумма всех депозитов на пользователя, который открыл хотя бы один депозит.

***H0: Средняя сумма всех депозитов на пользователя, который открыл хотя бы один депозит, одинакова и в контрольной, и в тестовой группе.***

***H1: Средняя сумма всех депозитов на пользователя, который открыл хотя бы один депозит, в тестовой группе отлична от средней суммы в контрольной.***

````python
#Проверим статистическую значимость изменения средней суммы всех депозитов на пользователя, который открыл хотя бы один депозит
avg_arpu_first = df_abt[df_abt['event_name'] == 'first_deposit'].groupby(['user_id', 'group'])['amount'].sum().reset_index().fillna(0)
control = avg_arpu_first[avg_arpu_first['group'] == 'control']['amount']
test = avg_arpu_first[avg_arpu_first['group'] == 'test']['amount']
_, p_value = ttest_ind(test, control, equal_var = False)
if p_value < 0.05:
    display(f'Значение p-value = {round(p_value, 4)}. Подтверждается альтернативная гипотеза. Изменение средней суммы всех депозитов на пользователя, который открыл хотя бы один депозит статистически значимы')
else:
    display(f'Значение p-value = {round(p_value, 4)}. Подтверждается нулевая гипотеза. Изменение средней суммы всех депозитов на пользователя, который открыл хотя бы один депозит статистически незначимы')
````
